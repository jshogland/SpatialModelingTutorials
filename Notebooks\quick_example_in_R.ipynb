**Here is a quick linear modleing example.**


---



Lets make a dataset with 10 observations and add some random noise.

1. The dataframe will be called df
2. The response variable is called y
3. The predictor variable is called x

Now let's look at our system




```r
library(ggplot2)
print(version)

df = data.frame(x=seq(10),y=rnorm(10,seq(10),1))
```

Now let's look at our data

```r
names(df)
nrow(df)
head(df)
summary(df)
```

From the summary statement we can get minimums, maximums, qluantiles etc.

Now let's run a linear regression and plot the results.

```r
mdl = lm(y~x,data=df)
summary(mdl)
plot(df$x,df$y)
lines(df$x,predict(mdl),col="Blue")
```

Let's make a prettier plot with ggplot.

```r
ggplot(df,aes(x,y))+geom_point()+geom_line(y=predict(mdl),col="Blue")
```

Let's check plot diagnostics

```r
plot(mdl)
```
